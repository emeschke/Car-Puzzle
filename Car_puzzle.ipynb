```python
import itertools
```

```python
cards = [0,1,2,3,4]
```

```python
all_cards = list(itertools.permutations(cards))
```

```python
def diffset(tup):
    diffs=[0]*4
    for i in range(4):
            diffs[i] = tup[i+1] - tup[0]
    return diffs
```

```python
del key_val
key_val = {}
```

```python
for i in range(len(all_cards)): 
    ds = tuple(diffset(all_cards[i]))
    print ds[0:3]
    if ds[0:3] in key_val.keys():
        key_val[ds[0:3]].append(all_cards[i])
        #key_val[ds[0:3]].append[all_cards[i]]
        print 'appended'
    else:
        key_val[ds[0:3]] = [all_cards[i]]
        print 'new'
    
```

```python
for i in key_val.keys():
    print i
    counter = 0
    forward_ev = 0
    stop_ev = 0
    for j in key_val[i]:
        counter += 1
        forward_ev += j[4]
        stop_ev += j[3]
    print 'Forward EV = ' + str(1.0*forward_ev/counter)
    print 'Stop EV = ' + str(1.0*stop_ev/counter)
    print 'Best EV = ' + str(min(1.0*forward_ev/counter, 1.0*stop_ev/counter))
    print 'Count is ' + str(counter)
```
