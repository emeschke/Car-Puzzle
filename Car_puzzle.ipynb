```python
import itertools
```

```python
#Define set of cards, [n, n+100, n+200, n+300, n+400]
cards = [0,1,2,3,4]
```

```python
#Get the list of all possible 5 card combos (120)
all_cards = list(itertools.permutations(cards))
len(all_cards)
```

```python
#Function that takes a set of cards and finds the difference sequence by subtracting off the first number.
def diffset(tup):
    diffs=[0]*4
    for i in range(4):
            diffs[i] = tup[i+1] - tup[0]
    return diffs
```

```python
#These will contain a list of all the possible combinations of cards indexed by their difference from the first value.
#This replicates what the customer sees as he/she engages in negotiations.
card_set_dict3 = {}
card_set_dict2 = {}
card_set_dict1 = {}
key_val = {}
```

```python
#Populate the card set dictionary with three differences.
for i in range(len(all_cards)): 
    ds = tuple(diffset(all_cards[i]))
    #print ds[0:3]
    if ds[0:3] in card_set_dict3.keys():
        card_set_dict3[ds[0:3]].append(all_cards[i])
        #print 'appended'
    else:
        card_set_dict3[ds[0:3]] = [all_cards[i]]
        #print 'new'
    
```

```python
#Populate the card set dictionary with two differences.
for i in range(len(all_cards)): 
    ds = tuple(diffset(all_cards[i]))
    #print ds[0:2]
    if ds[0:2] in card_set_dict2.keys():
        card_set_dict2[ds[0:2]].append(all_cards[i])
        #print 'appended'
    else:
        card_set_dict2[ds[0:2]] = [all_cards[i]]
        #print 'new'
```

```python
#Populate the card set dictionary with two differences.
for i in range(len(all_cards)): 
    ds = tuple(diffset(all_cards[i]))
    #print ds[0:1]
    if ds[0:1] in card_set_dict1.keys():
        card_set_dict1[ds[0:1]].append(all_cards[i])
        #print 'appended'
    else:
        card_set_dict1[ds[0:1]] = [all_cards[i]]
        #print 'new'
```

```python
#Initialize dictionaries that hold the minimum ev (stopping or forward) for each difference set and the count of
#outcomes that is in that set.  This provides the basis for the higher round's ev calcuation regarding stopping.
min_ev_count3 = {}
min_ev_count2 = {}
min_ev_count1 = {}
```

```python
#For the set of differences where four cards have been seen.
#Get the forward expected value (of continuing to the end in this case) and the value of stopping (the average 
#of the possible cards in the difference set.)
for i in card_set_dict3.keys():
    print i
    counter = 0
    forward_ev = 0
    stop_ev = 0
    for j in card_set_dict3[i]:
        print j
        counter += 1
        forward_ev += j[4]
        stop_ev += j[3]
    print 'Forward EV = ' + str(1.0*forward_ev/counter)
    print 'Stop EV = ' + str(1.0*stop_ev/counter)
    print 'Best EV = ' + str(min(1.0*forward_ev/counter, 1.0*stop_ev/counter))
    min_ev_count3[i] = (min(1.0*forward_ev/counter, 1.0*stop_ev/counter), counter)
    print 'Count is ' + str(counter)
```

```python
#For the set of differences where three cards have been seen.
#Get the forward expected value (of continuing to the end in this case) and the value of stopping (the average 
#of the possible cards in the difference set.)
for i in card_set_dict2.keys():
    print i
    counter = 0
    ev_count = 0
    forward_ev = 0
    stop_ev = 0
    #Stop EV
    for j in card_set_dict2[i]:
        print j
        counter += 1
        #forward_ev += j[3]
        stop_ev += j[2]
    #Forward EV of continuing
    for k in min_ev_count3.keys():
        if k[0:2] == i:
            print str(k[0:2]) + ' ' + str(i)
            #Expected value for that 
            forward_ev += min_ev_count3[k][0]*min_ev_count3[k][1]
            ev_count += min_ev_count3[k][1] 
    
    print 'Forward EV = ' + str(1.0*forward_ev/ev_count)
    print 'Stop EV = ' + str(1.0*stop_ev/counter)
    print 'Best EV = ' + str(min(1.0*forward_ev/ev_count, 1.0*stop_ev/counter))
    min_ev_count2[i] = (min(1.0*forward_ev/counter, 1.0*stop_ev/counter), counter)
    print 'Count is ' + str(counter)
    print 'EV count is ' + str(ev_count)
```
